```python
w2v_vectors = gensim.downloader.load('word2vec-google-news-300')
```

```python
compare_words = ['say', 'said', 'tell', 'told', 'state', 'stated', 'tree']

distances = dict(zip(compare_words, w2v_vectors.distances('say', compare_words)))

distances



```
